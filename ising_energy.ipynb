```python
import matplotlib.pyplot as plt

from isingmc import get_ising2D_sampler
```

```python
T = 2
L = 6

sampler = get_ising2D_sampler(beta=1/T, lattice_scale=L)
sample_energy = sampler.sample(
    nsamples=int(1e5), sample_interval=L**2, progress_bar_on=True)
```

```python
plt.plot(sample_energy)
plt.xlabel('N')
plt.ylabel('energy')
plt.title(r'Energy sample ($T=2$)')
plt.savefig('energy_singleT.png', bbox_inches='tight')
```

```python
energy_error=sampler.error_binning(sample_energy)
plt.plot(energy_error)
plt.xlabel(r'$l$')
plt.ylabel(r'$\sigma(E)$')
plt.title(r'Energy error ($T=2$)')
plt.savefig('error-l_singleT.png', bbox_inches='tight')
```
